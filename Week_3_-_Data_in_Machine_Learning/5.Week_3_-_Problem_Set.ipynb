__1. In a markdown block, describe the role of training testing data in detecting if a machine learning model is overfitting or underfitting.__

In machine learning, training testing data is used in detecting whether a model is overfitting or underfitting by providing a reference point with which to compare the model's predicted data. If the model does well in training with low cost predictions but has high cost with testing data then that model is likely overfitting. If the model does poorly in both training and testing then that model is likely underfitting.

__3. In a markdown box, explain bias and variance refering to the below model.__ 

```python
fig, ax = plt.subplots()

x = x + np.random.uniform(-3, 3, size=x.shape)
y = y + np.random.uniform(-5, 5, size=y.shape)
ax.scatter(x, y, c='b', label='Training & Testing data')
x = np.linspace(13, -13, 100)
y = -x**2 + 4*x + 2  
ax.plot(x, y, c='r', label='ML Model')
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_yticks([])
ax.set_yticklabels([])
plt.legend()
plt.show()
```

Bias refers to error in ML models that stems from oversimplistic modeling between the features and targets (underfitting). Variance refers to error in ML models that stems from overly complex modeling where the model will shape to the noise in the data (overfitting). The above model has fairly low bias and variance, meaning the model has good fitting.
